```python
import pandas

from bokeh.layouts import column
from bokeh.models import ColumnDataSource,RangeTool
from bokeh.plotting import figure,output_file,show
```

```python
df = pandas.read_html('https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20190220&end=20190320')
```

```python
df
```

在這個網頁裡面有兩個表格，所以在 df 回傳回來的值就會有兩個，但只要第一個表格，所以要用 index[0] 的方式，來回傳第一個表格。

此外，第一個表格的時間，是從 3/20 到 2/20，和所想要的時間順序是相反的，所以要用 [::-1] 的方式，全選表格後，進行反轉。

```python
df = pandas.read_html('https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20190220&end=20190320')[0][::-1]
```

```python
df
```

```python
df['Date'] = pandas.to_datetime(df['Date'])
```

```python
df
```

將原本輸出的 Date 時間，轉變為適當的時間格式，並會直接取代掉原本在 df 的 Date。

```python
dates = df['Date'].to_numpy(dtype = 'datetime64[D]')
```

將適當的時間格式，轉變為 numpy 的時間格式，在datetime64[] 裡面可以放入其它字元，就會顯示出不同的結果。

1. s 會直接顯示到秒。
2. m 會直接顯示到分。
3. h 會直接顯示到時。
4. D 會直接顯示到日。
5. M 會直接顯示到月。
6. Y 會直接顯示到年。

在秒，分，時這三個，都要用小寫，在年月日這三個，都要用大寫。

```python
dates
```

```python
source = ColumnDataSource(data = dict(data = dates, close = list(df["Close**"])))
```

對應到資料，在 data 的部分，就是對應到日期，在 close 的部分，就是對應到表格內的 close** 這部分。

```python
# 高度是 300
# 寬度是 1200
# 將預設的工具欄隱藏
# x軸資料為datetime
# x軸位置為上方（預設為下方）
# 背景填滿色為 #efefef
# 在圖表製作完後，x軸顯示預設的範圍為 2019-03-04 到 2019-03-12

p = figure(plot_height = 300, plot_width = 1200, tools = '', toolbar_location = None,x_axis_type ='datetime',x_axis_location = 'above',background_fill_color = '#efefef',x_range = (dates[12],dates[20]))
```

```python
p.line('data','close',source = source)
```

```python
# y軸圖例顯示為 price
p.yaxis.axis_label = 'Price'
```

```python
select = figure(title = 'Drag the middle and edges of selection box to change the range above',plot_height = 130,plot_width = 1200,y_range = p.y_range,x_axis_type = 'datetime',y_axis_type = None,tools = '',toolbar_location = None,background_fill_color = '#efefef')
```

```python
range_tool = RangeTool(x_range = p.x_range)
```

```python
range_tool.overlay.fill_color = 'navy'
range_tool.overlay.fill_alpha = 0.2
```

```python
select.line('data','close',source = source)
```

```python
select.ygrid.grid_line_color = None
```

```python
select.add_tools(range_tool)
```

```python
# 可以有多個動作，就像是選取起點跟終點
select.toolbar.active_multi = range_tool
```

```python
output_file('btc_range.html',title = 'Bitcoin Price Chart')
```

```python
# 將兩個圖表，當作是欄位，讓他們上面對齊，並顯示出來
show(column(p, select))
```
