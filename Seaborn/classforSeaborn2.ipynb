# Categorical Plots

```python
import seaborn as sns
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### 1. barplot can allows you to aggregate the categorical data and by default that's the mean

```python
# barplot(長條圖)
# 可以針對非數值欄位進行加總，並預設為算平均
# 可以將 x 和 y 進行顛倒

sns.barplot(x = 'sex', y = 'total_bill', data = tips)
```

```python
sns.barplot(x = 'total_bill', y = 'sex', data = tips)
```

```python
# barplot(長條圖)
# estimator can pass in whatever estimator function

sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std)
```

### 2. countplot can estimator is explicitly counting the number of occurrences

```python
# y軸是內建的，會是 count
# count 就是計數

sns.countplot(x = 'sex', data = tips)
```

### 3. boxplot can show the distribution of categorical data

```python
# boxplot(箱形圖)
# 顯示資料的分配情況，以四分位數的方式呈現
# 在 box 外的，就是離群值

sns.boxplot(x = 'day', y = 'total_bill', data = tips)
```

```python
# boxplot(箱形圖)
# 顯示資料的分配情況，以四分位數的方式呈現
# 在 box 外的，就是離群值
# 加上 hue，就可以將資料再進行分析
# 不單單只有呈現不同天的 total_bill，還有呈現 yes or no smoker

sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
```

### 4. violinplot very similar role to box plot

```python
# violinplot(小提琴圖)
# 可以看到資料密度及分佈的位置
# 空白圈圈是中位數
# 黑線代表信賴區間

sns.violinplot(x = 'day' , y = 'total_bill', data = tips)
```

```python
# violinplot(小提琴圖)
# 可以看到資料密度及分佈的位置
# 空白圈圈是中位數
# 黑線代表信賴區間

sns.violinplot(x = 'day' , y = 'total_bill', data = tips, hue = 'sex')
```

```python
# violinplot(小提琴圖)
# 可以看到資料密度及分佈的位置
# 空白圈圈是中位數
# 黑線代表信賴區間
# 可以將不同的 sex 放在一起
# 直接比較兩種不同 sex 之間的分配

sns.violinplot(x = 'day' , y = 'total_bill', data = tips, hue = 'sex', split = True)
```

### 5. stripplot draw a scatterplot where one variable is catagorical

```python
# stripplot(按照不同變數，繪製散佈圖)
# 可以看到有多少點是相互堆疊再一起

sns.stripplot(x = 'day', y = 'total_bill' , data = tips)
```

```python
# stripplot(按照不同變數，繪製散佈圖)
# jitter 是將堆疊的點，可以進行分開，如上圖所示

sns.stripplot(x = 'day', y = 'total_bill' , data = tips, jitter = False)
```

```python
# stripplot(按照不同變數，繪製散佈圖)
# jitter 是將堆疊的點，可以進行分開，如上圖所示
# hue 將資料區分為不同性別

sns.stripplot(x = 'day', y = 'total_bill' , data = tips, hue = 'sex')
```

```python
# stripplot(按照不同變數，繪製散佈圖)
# jitter 是將堆疊的點，可以進行分開，如上圖所示
# hue 將資料區分為不同性別
# dodge 是將不同性別的資料分開

sns.stripplot(x = 'day', y = 'total_bill' , data = tips, hue = 'sex', dodge = True)
```

### 6. swarmplot just like the stripplot except the points are adjusted so that they don't overlap

```python
# swarmplot(蜂群圖)
# 讓 point 不會互相堆疊
# 觀察這些 point 的分佈
# combination the stripplot and violinplot

sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
```

```python
# 可以將兩者在合併
# 可以顯示出 KDE

sns.violinplot(x = 'day' , y = 'total_bill', data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, color = 'black')
```

### 7. catplot 

```python
# 因子圖
# 可以利用 kind 讓 catplot 變成不同的形狀

sns.catplot(x = 'day' , y = 'total_bill', data = tips, kind = 'bar')
```
