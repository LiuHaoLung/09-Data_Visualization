```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

可以在 jupyter notebook 看到所創建的 plot。如果不是用 jupyter notebook，則需要打 <code>plot.show()</code>。

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# functional

plt.plot(x,y)
```

```python
plt.plot(x,y,'r')
```

<code>r</code>換顏色。

```python
plt.plot(x,y)
# x 軸名稱
plt.xlabel('X Label')
# y 軸名稱
plt.ylabel('Y Label')
# 圖表名稱
plt.title('Title')
```

```python
# 第一個是 column 的數量
# 第二個是 row 的數量
# 第三個是圖表的參照
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
## OO

# create a object
fig = plt.figure()

# left,bottom,width,height,use list,between 0 and 1
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)

# 設定 x軸
axes.set_xlabel('X Label')

# 設定 y軸
axes.set_ylabel('Y Label')

# 設定圖表名稱
axes.set_title('Title')
```

```python
fig = plt.figure()

# left,bottom,width,height,use list,between 0 and 1
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

# left,bottom,width,height,use list,between 0 and 1
# left 指的是原點的位置
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.16,0.4,0.3])
```

```python
fig = plt.figure()

# left,bottom,width,height,use list,between 0 and 1
# left 指的是原點的位置
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('Larger')
axes2.set_title('Smaller')
```
