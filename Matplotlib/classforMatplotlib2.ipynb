```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
# subplots 可以自動的做到呈現 axes1,axes2 
# 並且可以自訂欄列要多少
fig,axes = plt.subplots(nrows = 1,ncols = 2)
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)

# 可以用 index 的方式，來指定哪個圖表要呈現資料
axes[0].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)

# 可以用 index 的方式，來指定哪個圖表要呈現資料
# 並可以在不同 index ，設定不同的圖表名稱
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

# 可以解決不同圖表覆蓋的問題
plt.tight_layout()
```

```python
# width,height,use tuple,dpi
fig = plt.figure(figsize = (3,2),dpi = 100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# width,height,use tuple,dpi
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (3,2),dpi = 500)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()

fig.savefig('my_plot.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
ax.plot(y,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# set legend use label argument
ax.plot(x,x ** 2,label = 'x Squared')
ax.plot(x,x ** 3,label = 'x Cubed')

# and use ax.legend() can do that
# many location can use,just add loc = num
# can use tuple to decide the legend location
ax.legend()
```
