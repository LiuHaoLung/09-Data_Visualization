# Regression Plots

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### 1. lmplot can display the linear model

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex')
```

```python
# markers can change the icon in the plot
# have to sex so the markers need to pass in two different icon

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o','v'])
```

```python
# use scatter_kws can change the markers size
# when change the size it should be use dict

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o','v'],
          scatter_kws = {'s' : 100})
```

```python
# use col can separated the sex to 2 columns

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'sex')
```

```python
# use col can separated the sex to 2 columns
# use row can separated the time to 2 rows

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'sex', row = 'time')
```

```python
# use col can separated the sex to 2 columns
# use row can separated the time to 2 rows
# use hue can let every plot have different data

sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'day', row = 'time', hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'day', hue = 'sex')
```

```python
# aspect and height can change plots size
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'day', hue = 'sex',
          aspect = 0.6, height = 8)
```
