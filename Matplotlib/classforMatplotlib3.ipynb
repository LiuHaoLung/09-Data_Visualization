```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# put the color name
# or RGB hex color code
ax.plot(x,y,color = 'orange')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# put the color name
# or RGB hex color code
ax.plot(x,y,color = '#FF8C00')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# change the line width use linewidth,default is 1,can just put lw
ax.plot(x,y,color = '#FF8C00',linewidth = 2)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# alpha can transparency
ax.plot(x,y,color = '#FF8C00',linewidth = 2,alpha = 0.5)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# linestyle can change the line looks like,can just put ls
ax.plot(x,y,color = '#FF8C00',linewidth = 2,alpha = 0.5,linestyle = '--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# marker can specify the data is
ax.plot(x,y,color = '#FF8C00',lw = 2,alpha = 0.5,ls = '--',marker = 'o')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# marker can specify the data is
# markersize can change the marker size
ax.plot(x,y,color = '#FF8C00',lw = 2,alpha = 0.5,ls = '--',marker = 'o',markersize = 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# marker can specify the data is
# markersize can change the marker size
# markerfacecolor can change the marker color
# markeredgewidth can change the marker edge size
# markeredgecolor can change the marker edge color
ax.plot(x,y,color = '#FF8C00',lw = 2,alpha = 0.5,ls = '--',marker = 'o',
        markersize = 10,markerfacecolor = 'red',markeredgewidth = 5,
       markeredgecolor = 'black')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = '#FF8C00',lw = 2,alpha = 0.5,ls = '--')

# set the x axis lower is 0 and upper is 1
ax.set_xlim([0,1])

# set the y axis lower is 0 and upper is 2
ax.set_ylim([0,2])
```
