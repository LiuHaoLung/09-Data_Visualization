# Style and color

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### 1. set_style

```python
sns.countplot(x = 'sex',data = tips)
```

```python
# default the background is white

sns.set_style('white')
sns.countplot(x = 'sex',data = tips)
```

```python
# ticks basically get these ticks at the edge of the box

sns.set_style('ticks')
sns.countplot(x = 'sex',data = tips)
```

```python
# background have the lines and the color is black

sns.set_style('darkgrid')
sns.countplot(x = 'sex',data = tips)
```

```python
# background have the lines and the color is white

sns.set_style('whitegrid')
sns.countplot(x = 'sex',data = tips)
```

```python
# despine can remove the bottom and left spine
# default left and bottom are false

sns.set_style('ticks')
sns.countplot(x = 'sex',data = tips)
sns.despine()
```

```python
# despine can remove the bottom and left spine
# default left and bottom are false
# if the left and bottom are all true then can remove the spine

sns.set_style('ticks')
sns.countplot(x = 'sex',data = tips)
sns.despine(left = True, bottom = True)
```

```python
# matplotlib set the size for the seaborn

plt.figure(figsize = (12,3))
sns.countplot(x = 'sex',data = tips)
```

```python
# because the plot is going to put this on poster

sns.set_context('poster')
sns.countplot(x = 'sex',data = tips)
```

```python
# because the plot is going to put this on poster
# font_scale can change the font size

sns.set_context('poster', font_scale = 2)
sns.countplot(x = 'sex',data = tips)
```

```python
# because the plot is going to put this on notebook

sns.set_context('notebook')
sns.countplot(x = 'sex',data = tips)
```

### 2. palettes and colors

```python
sns.lmplot(x = 'total_bill',y = 'tip', data = tips)
```

```python
# palette can find in the https://matplotlib.org/examples/color/colormaps_reference.html

sns.lmplot(x = 'total_bill',y = 'tip', data = tips,hue = 'sex', palette = 'coolwarm')
```

```python
# palette can find in the https://matplotlib.org/examples/color/colormaps_reference.html

sns.lmplot(x = 'total_bill',y = 'tip', data = tips,hue = 'sex', palette = 'seismic')
```
